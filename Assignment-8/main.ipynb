 Apply WordNet-based Word Sense Disambiguation to improve meaning interpretation in 
ambiguous sentences. 

```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.wsd import lesk

nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
```

```python
sent1 = "I went to the bank to deposit money"
sent2 = "The children are playing near the river bank"
```

```python
sense1 = lesk(nltk.word_tokenize(sent1), 'bank')
sense2 = lesk(nltk.word_tokenize(sent2), 'bank')

print("Sentence 1 Sense:", sense1, "-", sense1.definition())
print("Sentence 2 Sense:", sense2, "-", sense2.definition())
```

```python
sentence = "He sat on the chair and cracked a joke"
sense = lesk(nltk.word_tokenize(sentence), 'cracked')
print(sense, "-", sense.definition())
```
