Identifying Semantic Relationships using WordNet

```python
import nltk
from nltk.corpus import wordnet as wn

nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
words = ["happy", "good", "car"]
```

```python
synonyms = set()

for syn in wn.synsets("happy"):
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())

print("Synonyms of 'happy':", synonyms)
```

```python
antonyms = set()

for syn in wn.synsets("happy"):
    for lemma in syn.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

print("Antonyms of 'happy':", antonyms)
```

```python
synsets = wn.synsets("car")
for syn in synsets[:1]:   # using first sense
    hypernyms = syn.hypernyms()
    print("Hypernyms of 'car':")
    for h in hypernyms:
        print(" -", h.name())
```

```python
hyponyms = syn.hyponyms()
print("Hyponyms of 'car':")
for h in hyponyms[:5]:
    print(" -", h.name())
```
