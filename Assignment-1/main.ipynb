```python
text = "Hii this is Mansi! And I am studying product management and its movtivating :) #aspiringPM"
```

Tokenization using NLTK

(a) Whitespace Tokenization

```python
from nltk.tokenize import WhitespaceTokenizer

wt = WhitespaceTokenizer()
tokens_whitespace = wt.tokenize(text)

print("Whitespace Tokenization:", tokens_whitespace)
```

(b) Punctuation-based Tokenization

```python
from nltk.tokenize import wordpunct_tokenize

tokens_punct = wordpunct_tokenize(text)
print("Punctuation-based Tokenization:", tokens_punct)
```

(c) Treebank Tokenization

```python
from nltk.tokenize import TreebankWordTokenizer

treebank = TreebankWordTokenizer()
tokens_treebank = treebank.tokenize(text)

print("Treebank Tokenization:", tokens_treebank)
```

(d) Tweet Tokenization

```python
from nltk.tokenize import TweetTokenizer

tweet_tokenizer = TweetTokenizer()
tokens_tweet = tweet_tokenizer.tokenize(text)

print("Tweet Tokenization:", tokens_tweet)
```

(e) Multi-Word Expression (MWE) Tokenization

```python
from nltk.tokenize import MWETokenizer

mwe_tokenizer = MWETokenizer([('product', 'management'),
                              ('aspiring', 'pm')], separator='_')

text_mwe = "I love product management and I am an aspiring pm"

tokens_mwe = mwe_tokenizer.tokenize(text_mwe.split())

print("MWE Tokenization:", tokens_mwe)
```

Stemming

(a) Porter Stemmer

```python
from nltk.stem import PorterStemmer

porter = PorterStemmer()
porter_stems = [porter.stem(word) for word in tokens_treebank]

print("Porter Stemmer:", porter_stems)
```

(b) Snowball Stemmer

```python
from nltk.stem import SnowballStemmer

snowball = SnowballStemmer("english")
snowball_stems = [snowball.stem(word) for word in tokens_treebank]

print("Snowball Stemmer:", snowball_stems)
```

Lemmatization

WordNet Lemmatizer

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmas = [lemmatizer.lemmatize(word.lower()) for word in tokens_treebank]

print("Lemmatization:", lemmas)
```
